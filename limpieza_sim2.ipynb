```python
sust = ' IR Blaster'
# Recorrer la columna en busca de la palabra
for indice, valor in df['Vo5G'].items():
    if sust in valor and df.at[indice, 'extra_conecctions'] == 'No data':
            df.at[indice, 'extra_conecctions'] = 'IR Blaster'
            df.at[indice, 'Vo5G'] = 'No data'
```

```python
sust = ' NFC'
# Recorrer la columna en busca de la palabra
for indice, valor in df['Vo5G'].items():
    if sust in valor and df.at[indice, 'NFC'] == ' IR Blaster' and df.at[indice, 'extra_conecctions'] == 'No data':
            df.at[indice, 'extra_conecctions'] = 'IR Blaster'
            df.at[indice, 'NFC'] = 'NFC'
            df.at[indice, 'Vo5G'] = 'No data'
```

```python
sust = ' IR Blaster'
# Recorrer la columna en busca de la palabra
for indice, valor in df['Vo5G'].items():
    if sust in valor and df.at[indice, 'NFC'] == 'No data' and df.at[indice, 'extra_conecctions'] == ' NFC':
            df.at[indice, 'extra_conecctions'] = 'IR Blaster'
            df.at[indice, 'NFC'] = 'NFC'
            df.at[indice, 'Vo5G'] = 'No data'
```

```python
# Recorrer la columna en busca de la palabra
for indice, valor in df['additional_data'].items():
    if valor != 'No data' and  df.at[indice, 'extra_conecctions'] == 'No data':
            df.at[indice, 'extra_conecctions'] = df.at[indice, 'additional_data']
            df.at[indice, 'additional_data'] = 'No data'
            
```

```python
# Recorrer la columna en busca de la palabra
for indice, valor in df['additional_data'].items():
    if valor != 'No data':
            df.at[indice, 'extra_conecctions'] = df.at[indice, 'extra_conecctions']+ ', '+ df.at[indice, 'additional_data']
            df.at[indice, 'additional_data'] = 'No data'
```
