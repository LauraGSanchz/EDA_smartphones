## Análisis de cardinalidad


```python
cardinalidad = pd.DataFrame(columns=['cardinalidad','porcentaje_cardinalidad','tipo_de_dato','valores_unicos'],  index=smartphones_clean.columns)
```

```python
cardinalidad['cardinalidad']= [smartphones_clean[col].nunique() for col in smartphones_clean.columns]
cardinalidad['porcentaje_cardinalidad'] = (cardinalidad['cardinalidad']/len(smartphones_clean)*100).round(2)
cardinalidad['tipo_de_dato'] = smartphones_clean.dtypes
cardinalidad['valores_unicos'] = [smartphones_clean[col].unique() for col in smartphones_clean.columns]
```
